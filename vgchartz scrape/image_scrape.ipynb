## Image Scrape

VGChartz provide a great database of image box arts for the games

If you selected to scrape the URL of the image like in this repo, you can use this second notebook to download the images

The images will be downloaded to a folder with the name of the console.






```python
import pandas as pd

# Specify the file path
file_path = 'scrapped_data.xlsx'

# Read the Excel file
df = pd.read_excel(file_path)

# Display the data
df.head()
#open the game_sales_cleaned.xlsx
```

```python
import requests
import os

# Create a folder to save the images
folder_path = 'image_folder'
os.makedirs(folder_path, exist_ok=True)

# Iterate over each row in the dataframe
for index, row in df.iterrows():
    image_url = row['Image_URL']
    system_full = row['System Full']
    
    # Create a folder for each system
    system_folder = os.path.join(folder_path, system_full)
    os.makedirs(system_folder, exist_ok=True)
    
    # Download the image and save it to the system folder
    response = requests.get(image_url)
    image_path = os.path.join(system_folder, f'image_{index}.jpg')
    with open(image_path, 'wb') as f:
        f.write(response.content)
```
