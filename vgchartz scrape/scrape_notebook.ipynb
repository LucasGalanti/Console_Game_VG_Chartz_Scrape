```python
def open_url(system_desired):
    base_url = "https://www.vgchartz.com/games/games.php?"
    params = {
        'name': '',
        'keyword': '',
        'console': '',  # Placeholder for the console name
        'region': 'All',
        'developer': '',
        'publisher': '',
        'goty_year': '',
        'genre': '',
        'boxart': 'Both',
        'banner': 'Both',
        'ownership': 'Both',
        'showmultiplat': 'Yes',
        'results': '100000',         ## put a huge value here to get all the results in one page
        'order': 'TotalShipped',
        'showtotalsales': '1',
        'showpublisher': '1',
        'showvgchartzscore': '0',
        'shownasales': '0',
        'showdeveloper': '1',
        'showcriticscore': '0',
        'showpalsales': '0',
        'showreleasedate': '1',
        'showuserscore': '0',
        'showjapansales': '0',
        'showlastupdate': '0',
        'showothersales': '0',
        'showshipped': '1'
    }

    params['console'] = system_desired  # Set the console parameter
    console_url = base_url + '&'.join([f"{k}={v}" for k, v in params.items()])
    response = requests.get(console_url)
    html = response.content
    soup = BeautifulSoup(html, 'html.parser')
    games_table_in = soup.find('div', id='generalBody')
    rows_in = games_table_in.find_all('tr', style=True)  # Retrieve all the rows first to count them

    return rows_in


```

```python
games_data = []
# Use tqdm to wrap the iteration and provide a progress bar

for console in tqdm(desired_consoles.keys(), desc="looking in systems", unit = "system"):  # Fix: Call the keys() method on desired_consoles
    if desired_consoles[console][-1] == 1:
        rows = open_url(desired_consoles[console][0])
        for game_row in tqdm(rows, desc="Scraping games for "+ console , unit="row"):
            cells = game_row.find_all('td')
            system_image = cells[3].find('img')  # Find the <img> tag within the System cell
            system = system_image['alt'] if system_image and 'alt' in system_image.attrs else 'No system info'  # Get the alt text or a default
            game_image_url = cells[1].find('img')['src'] if cells[1].find('img') else 'N/A'

            if len(cells) > 1:  # This checks if the row is not a header or empty
                game_info = {
                    'Game Name': cells[2].get_text(strip=True),
                    'Image_URL': "https://www.vgchartz.com"+game_image_url,
                    'System': system,
                    'Publisher': cells[4].get_text(strip=True),
                    'Developer': cells[5].get_text(strip=True),
                    'Total Shipped': cells[6].get_text(strip=True),
                    'Total Sold': cells[7].get_text(strip=True),  # Adjust if the index is different
                    'Release Date': cells[8].get_text(strip=True)
                }
                games_data.append(game_info)


```

## Data Cleaning


```python
df = pd.DataFrame(games_data)
```

```python
# Edit date column
df['Release Date'] = df['Release Date'].str.replace('N/A', '')
df['Release Date'] = pd.to_datetime(df['Release Date'], format='mixed', dayfirst=True).dt.strftime('%d/%m/%Y')
```

```python
df['Game Name'] = df['Game Name'].str.replace('Read the review', '', regex=False)
```

```python



df['Total Shipped'] = df['Total Shipped'].str.replace('m', '').replace('N/A', '0').astype(float)
df['Total Sold'] = df['Total Sold'].str.replace('m', '').replace('N/A', '0').astype(float)
df['Units(m)'] = df['Total Shipped']+df['Total Sold']
```

```python
reverse_dict = {v[0]: k for k, v in desired_consoles.items()}
```

```python
df['System Full'] = df['System'].map(reverse_dict)
```

```python
df['System Full'] = df['System Full'].str.replace('Atari 5200', 'rep').replace('Atari 7800','rep').replace('rep', 'Atari 5200/7800')
df['System Full'] = df['System Full'].str.replace('Super Nintendo Entertainment System', 'Super Nintendo')
```

```python
df = df.drop(['Total Shipped', 'Total Sold', 'System'], axis=1)
```

```python
df = df.reindex(columns=['Game Name', 'System Full', 'Units(m)', 'Publisher', 'Developer','Image_URL', 'Release Date'])
```
