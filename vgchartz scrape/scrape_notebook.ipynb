```python
base_url = "https://www.vgchartz.com/games/games.php?"
params = {
    'name': '',
    'keyword': '',
    'console': '',  # Placeholder for the console name
    'region': 'All',
    'developer': '',
    'publisher': '',
    'goty_year': '',
    'genre': '',
    'boxart': 'Both',
    'banner': 'Both',
    'ownership': 'Both',
    'showmultiplat': 'Yes',
    'results': '10',
    'order': 'TotalShipped',
    'showtotalsales': '1',
    'showpublisher': '1',
    'showvgchartzscore': '0',
    'shownasales': '0',
    'showdeveloper': '1',
    'showcriticscore': '1',
    'showpalsales': '0',
    'showreleasedate': '1',
    'showuserscore': '0',
    'showjapansales': '0',
    'showlastupdate': '0',
    'showothersales': '0',
    'showshipped': '1'
}

# List of consoles you want to search for, and the corresponding value to be put in the URL
consoles_list = ['DS', 'PS4', 'XOne']

# List to hold the URLs
urls_for_consoles = []

# Generate a URL for each console in the list
for console in consoles_list:
    params['console'] = console  # Set the console parameter
    console_url = base_url + '&'.join([f"{k}={v}" for k, v in params.items()])
    urls_for_consoles.append(console_url)

```

```python
url = "https://www.vgchartz.com/games/games.php?name=&keyword=&console=PS4&region=All&developer=&publisher=&goty_year=&genre=&boxart=Both&banner=Both&ownership=Both&showmultiplat=Yes&results=100000&order=TotalShipped&showtotalsales=0&showtotalsales=1&showpublisher=0&showpublisher=1&showvgchartzscore=0&shownasales=0&showdeveloper=0&showdeveloper=1&showcriticscore=0&showpalsales=0&showreleasedate=0&showreleasedate=1&showuserscore=0&showjapansales=0&showlastupdate=0&showothersales=0&showshipped=0&showshipped=1"

response = requests.get(url)
html = response.content
soup = BeautifulSoup(html, 'html.parser')

games_table = soup.find('div', id='generalBody')
rows = games_table.find_all('tr', style=True)  # Retrieve all the rows first to count them
```

```python
games_data = []
# Use tqdm to wrap the iteration and provide a progress bar
for game_row in tqdm(rows, desc="Scraping games", unit="row"):
    cells = game_row.find_all('td')
    system_image = cells[3].find('img')  # Find the <img> tag within the System cell
    system = system_image['alt'] if system_image and 'alt' in system_image.attrs else 'No system info'  # Get the alt text or a default
    game_image_url = cells[1].find('img')['src'] if cells[1].find('img') else 'N/A'

    if len(cells) > 1:  # This checks if the row is not a header or empty
        game_info = {
            'Game Name': cells[2].get_text(strip=True),
            'Image_URL': "https://www.vgchartz.com"+game_image_url,
            'System': system,
            'Publisher': cells[4].get_text(strip=True),
            'Developer': cells[5].get_text(strip=True),
            'Total Shipped': cells[6].get_text(strip=True),
            'Total Sold': cells[7].get_text(strip=True),  # Adjust if the index is different
            'Release Date': cells[8].get_text(strip=True)
        }
        games_data.append(game_info)

df = pd.DataFrame(games_data)
```

```python
df['Total Shipped'] = df['Total Shipped'].str.replace('m', '').replace('N/A', '0').astype(float)
df['Total Sold'] = df['Total Sold'].str.replace('m', '').replace('N/A', '0').astype(float)

df['Units(m)'] = df['Total Shipped']+df['Total Sold']

df = df.drop(['Total Shipped', 'Total Sold'], axis=1)

df.head()
```

```python
df['Release Date'] = pd.to_datetime(df['Release Date'], format='mixed', dayfirst=True)
```
